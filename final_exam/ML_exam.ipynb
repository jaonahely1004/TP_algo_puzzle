```
# Partie I : Dataset
import pandas as pd
import urllib.request
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import nest_asyncio
nest_asyncio.apply()
import threading
import numpy as np



# telechargement du dataset depuis le site UCI Machine Learning
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/tic-tac-toe/tic-tac-toe.data"
file_path = "tic_tac_toe.data"
urllib.request.urlretrieve(url, file_path)

# Nom des colonnes
columns = [
    "top-left", "top-middle", "top-right",
    "middle-left", "middle-middle", "middle-right",
    "bottom-left", "bottom-middle", "bottom-right",
    "class"
]

# Lire le fichier CSV
df = pd.read_csv(file_path, names=columns)

#afficher le dataset
pd.set_option('display.max_rows', None)
print("‚úÖ Affichage complet du dataset :")
display(df)

#Statistiques de classe
print("\nüìä R√©partition des classes :")
print(df["class"].value_counts())

#encodage des valeurs
mapping = {'x': 1, 'o': 2, 'b': 0}
for col in columns[:-1]:
    df[col] = df[col].map(mapping)

df['class'] = df['class'].map({'positive': 1, 'negative': 0})

print("\n‚úÖ Dataset encod√© pr√™t pour l'apprentissage :")
display(df)

# Partie II : Apprentissage

#Separation x (features) et y (cible)
X = df.drop("class", axis=1) # X colonne du plateau
y = df["class"] # y classe (posi/nega)

#Separation en donn√©e l'entrainement et le test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

#Creation du modele d'arbre de decision
model = DecisionTreeClassifier(random_state=42)

#Entrainer sur le donn√©es d'entrainement
model.fit(X_train, y_train)

# Pr√©diction
y_pred = model.predict(X_test)

#Affichage graphique
cm = confusion_matrix(y_test, y_pred)
labels = ['negative (0)' , 'positive (1)']
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d' , cmap='Blues', xticklabels=labels, yticklabels=labels)

plt.title(f"üßÆ Matrice de confusion - Pr√©cision : {accuracy_score(y_test, y_pred)*100:.2f}%")
plt.xlabel("Pr√©diction")
plt.ylabel("R√©el")
plt.tight_layout()
plt.show()

#Partie 3
import ipywidgets as widgets
from IPython.display import display

# Plateau initial
plateau = [0] * 9
symboles = {0: ' ', 1: '‚ùå', 2: '‚≠ï'}

# Colonnes du dataset (partie 1)
colonnes_plateau = [
    "top-left", "top-middle", "top-right",
    "middle-left", "middle-middle", "middle-right",
    "bottom-left", "bottom-middle", "bottom-right"
]

# Cr√©ation des boutons
boutons = [widgets.Button(description=' ', layout=widgets.Layout(width='70px', height='70px')) for _ in range(9)]
texte_resultat = widgets.HTML(value="<b>Votre tour (‚ùå)</b>")

# Fonction pour v√©rifier un gagnant
def verifier_gagnant(p, joueur):
    combinaisons = [
        [0,1,2], [3,4,5], [6,7,8],
        [0,3,6], [1,4,7], [2,5,8],
        [0,4,8], [2,4,6]
    ]
    return any(all(p[i] == joueur for i in ligne) for ligne in combinaisons)

# Fonction d'√©valuation combin√©e (Minimax + mod√®le ML)
def evaluer_position(p):
    if verifier_gagnant(p, 2):
        return 1
    elif verifier_gagnant(p, 1):
        return -1
    elif all(x != 0 for x in p):
        return 0

    # Position interm√©diaire : utiliser mod√®le ML comme heuristique
    df_test = pd.DataFrame([p], columns=colonnes_plateau)
    pred = model.predict(df_test)[0] #model = arbre de decision dans la partie 2
    # 0 = favorable IA, 1 = favorable humain
    return 0.5 if pred == 0 else -0.5

# Algorithme Minimax avec alpha-beta pruning
def minimax(p, is_maximizing, alpha, beta):
    score = evaluer_position(p)
    # Si score terminal (victoire/d√©faite/nul), on retourne
    if abs(score) == 1 or score == 0:
        return score

    if is_maximizing:
        max_eval = -float('inf')
        for i in range(9):
            if p[i] == 0:
                p[i] = 2
                eval = minimax(p, False, alpha, beta)
                p[i] = 0
                max_eval = max(max_eval, eval)
                alpha = max(alpha, eval)
                if beta <= alpha:
                    break
        return max_eval
    else:
        min_eval = float('inf')
        for i in range(9):
            if p[i] == 0:
                p[i] = 1
                eval = minimax(p, True, alpha, beta)
                p[i] = 0
                min_eval = min(min_eval, eval)
                beta = min(beta, eval)
                if beta <= alpha:
                    break
        return min_eval

# IA joue son coup optimal
def ia_jouer():
    global plateau
    meilleur_score = -float('inf')
    meilleur_coup = None

    for i in range(9):
        if plateau[i] == 0:
            plateau[i] = 2
            score = minimax(plateau, False, -float('inf'), float('inf'))
            plateau[i] = 0
            if score > meilleur_score:
                meilleur_score = score
                meilleur_coup = i

    if meilleur_coup is not None:
        plateau[meilleur_coup] = 2
        boutons[meilleur_coup].description = symboles[2]
        boutons[meilleur_coup].disabled = True

def desactiver_tous():
    for btn in boutons:
        btn.disabled = True

def on_click(index):
    def handler(btn):
        global plateau
        if plateau[index] == 0:
            plateau[index] = 1
            btn.description = symboles[1]
            btn.disabled = True

            if verifier_gagnant(plateau, 1):
                texte_resultat.value = "<b style='color:green;'>üéâ Vous avez gagn√© !</b>"
                desactiver_tous()
                return

            if all(x != 0 for x in plateau):
                texte_resultat.value = "<b style='color:orange;'>ü§ù Match nul !</b>"
                return

            ia_jouer()

            if verifier_gagnant(plateau, 2):
                texte_resultat.value = "<b style='color:red;'>üíª L‚ÄôIA a gagn√© !</b>"
                desactiver_tous()
            elif all(x != 0 for x in plateau):
                texte_resultat.value = "<b style='color:orange;'>ü§ù Match nul !</b>"
    return handler

def reset(_=None):
    global plateau
    plateau = [0] * 9
    texte_resultat.value = "<b>Votre tour (‚ùå)</b>"
    for i, btn in enumerate(boutons):
        btn.description = symboles[0]
        btn.disabled = False

for i, btn in enumerate(boutons):
    btn.on_click(on_click(i))

btn_reset = widgets.Button(description="üîÑ Rejouer", button_style='info')
btn_reset.on_click(reset)

grille = widgets.GridBox(
    children=boutons,
    layout=widgets.Layout(
        width='200px',
        grid_template_columns='60px 60px 60px',
        grid_template_rows='60px 60px 60px',
        grid_gap='5px'
    )
)

interface = widgets.VBox([texte_resultat, grille, btn_reset])
display(interface)
```
